Varianza y desviación estandar para conjutos de datos pares e impares

```python
import math
from prettytable import PrettyTable

class EstadisticasCalculadora:
    def __init__(self, datos):
        self.datos = datos
        self.datos_cuadrados = [x ** 2 for x in self.datos]
        self.ordenar_datos()

    def calcular_media(self):
        suma_datos = 0
        cantidad_datos = len(self.datos)
        iteraciones_media = []

        for i, dato in enumerate(self.datos, start=1):
            suma_datos += dato
            iteraciones_media.append([dato, suma_datos, i])

        media = suma_datos / cantidad_datos if cantidad_datos > 0 else None
        return media, iteraciones_media

    def ordenar_datos(self):
        self.datos_ordenados = sorted(self.datos)
        self.iteraciones_ordenamiento = []

        for i in range(len(self.datos_ordenados)):
            for j in range(0, len(self.datos_ordenados) - i - 1):
                if self.datos_ordenados[j] > self.datos_ordenados[j + 1]:
                    self.datos_ordenados[j], self.datos_ordenados[j + 1] = self.datos_ordenados[j + 1], self.datos_ordenados[j]

                self.iteraciones_ordenamiento.append(self.datos_ordenados.copy())

    def calcular_mediana(self):
        cantidad_datos = len(self.datos_ordenados)
        iteraciones_mediana = []

        if cantidad_datos == 0:
            return None, iteraciones_mediana  # Evitar el cálculo si no hay datos

        if cantidad_datos % 2 == 0:
            indice_medio_superior = cantidad_datos // 2
            indice_medio_inferior = indice_medio_superior - 1
            mediana = (self.datos_ordenados[indice_medio_inferior] + self.datos_ordenados[indice_medio_superior]) / 2
        else:
            indice_medio = cantidad_datos // 2
            mediana = self.datos_ordenados[indice_medio]

        return mediana, iteraciones_mediana

    def calcular_sumatoria_cuadrados(self):
        suma_cuadrados = 0
        iteraciones_sumatoria_cuadrados = []

        for i, cuadrado in enumerate(self.datos_cuadrados, start=1):
            suma_cuadrados += cuadrado
            iteraciones_sumatoria_cuadrados.append([i, cuadrado, suma_cuadrados])

        return suma_cuadrados, iteraciones_sumatoria_cuadrados

    def restar_media_a_datos(self):
        media = self.calcular_media()[0]

        if media is not None:
            datos_resta_media = [x - media for x in self.datos]
            iteraciones_resta_media = [[x, x - media] for x in self.datos]
            return datos_resta_media, iteraciones_resta_media
        else:
            return None, None  # No hay datos para restar la media

    def calcular_sumatoria_cuadrados_resta_media(self):
        datos_resta_media, _ = self.restar_media_a_datos()
        datos_cuadrados_resta_media = [x ** 2 for x in datos_resta_media]

        suma_cuadrados_resta_media = 0
        iteraciones_sumatoria_cuadrados_resta_media = []

        for i, cuadrado in enumerate(datos_cuadrados_resta_media, start=1):
            suma_cuadrados_resta_media += cuadrado
            iteraciones_sumatoria_cuadrados_resta_media.append([i, cuadrado, suma_cuadrados_resta_media])

        return suma_cuadrados_resta_media, iteraciones_sumatoria_cuadrados_resta_media


class DesviacionDeLaMedia:
    def __init__(self, estadisticas_calculadora):
        self.estadisticas_calculadora = estadisticas_calculadora
        self.sumatoria_cuadrados_resta_media, _ = self.estadisticas_calculadora.calcular_sumatoria_cuadrados_resta_media()
        self.numero_valores = len(self.estadisticas_calculadora.datos)

    def calcular_desviacion_media(self):
        if self.numero_valores > 1:
            desviacion_media = self.sumatoria_cuadrados_resta_media / (self.numero_valores - 1)
            return desviacion_media
        else:
            return None


class DesviacionEstandar:
    def __init__(self, desviacion_media_calculadora):
        self.desviacion_media_calculadora = desviacion_media_calculadora
        self.desviacion_media = self.desviacion_media_calculadora.calcular_desviacion_media()

    def calcular_desviacion_estandar(self):
        if self.desviacion_media is not None:
            desviacion_estandar = math.sqrt(self.desviacion_media)
            return desviacion_estandar
        else:
            return None


def imprimir_tabla_combinada(iteraciones_media, iteraciones_resta_media, iteraciones_sumatoria_cuadrados, iteraciones_sumatoria_cuadrados_resta_media, resultado_media):
    tabla_combinada = PrettyTable(["Dato", "Suma Acumulada", "Cantidad de Datos", "Resta de la Media", "Cuadrado", "Suma Acumulada Cuadrados", "Cuadrado - Media", "Suma Acumulada Cuadrados - Media"])

    if resultado_media is not None:
        sumas_acumuladas = [0] * (len(tabla_combinada.field_names) - 1)

        for iteracion_media, iteracion_resta_media, iteracion_sumatoria_cuadrados, iteracion_sumatoria_cuadrados_resta_media in zip(iteraciones_media, iteraciones_resta_media, iteraciones_sumatoria_cuadrados, iteraciones_sumatoria_cuadrados_resta_media):
            fila = iteracion_media + iteracion_resta_media[1:] + iteracion_sumatoria_cuadrados[1:] + [iteracion_resta_media[1] ** 2, iteracion_sumatoria_cuadrados[2] - resultado_media]
            tabla_combinada.add_row(fila)

            # Actualizar las sumas acumuladas
            sumas_acumuladas = [sumas_acumuladas[i] + fila[i + 1] for i in range(len(sumas_acumuladas))]

        # Agregar la fila de sumatorias
        sumatorias = ["Sumatoria"] + sumas_acumuladas
        tabla_combinada.add_row(sumatorias)

        print("Resultados combinados:")
        print(tabla_combinada)
    else:
        print("No hay datos para calcular la media.")


# Ejemplo de uso
if __name__ == "__main__":
    # Datos de ejemplo
    datos_ejemplo_par = [10, 15, 20, 25, 30, 35]
    datos_ejemplo_impar = [10, 15, 20, 25, 30]

    # Crear una instancia de la clase EstadisticasCalculadora para datos pares
    estadisticas_calculadora_par = EstadisticasCalculadora(datos=datos_ejemplo_par)

    # Crear una instancia de la clase EstadisticasCalculadora para datos impares
    estadisticas_calculadora_impar = EstadisticasCalculadora(datos=datos_ejemplo_impar)

    # Calcular la media para datos pares
    resultado_media_par, iteraciones_media_par = estadisticas_calculadora_par.calcular_media()

    # Calcular la media para datos impares
    resultado_media_impar, iteraciones_media_impar = estadisticas_calculadora_impar.calcular_media()

    # Restar la media a cada valor de datos_ejemplo e imprimir los resultados para datos pares
    datos_resta_media_par, iteraciones_resta_media_par = estadisticas_calculadora_par.restar_media_a_datos()

    # Restar la media a cada valor de datos_ejemplo e imprimir los resultados para datos impares
    datos_resta_media_impar, iteraciones_resta_media_impar = estadisticas_calculadora_impar.restar_media_a_datos()

    # Calcular la sumatoria de los cuadrados e imprimir el resultado y las iteraciones para datos pares
    resultado_sumatoria_cuadrados_par, iteraciones_sumatoria_cuadrados_par = estadisticas_calculadora_par.calcular_sumatoria_cuadrados()

    # Calcular la sumatoria de los cuadrados e imprimir el resultado y las iteraciones para datos impares
    resultado_sumatoria_cuadrados_impar, iteraciones_sumatoria_cuadrados_impar = estadisticas_calculadora_impar.calcular_sumatoria_cuadrados()

    # Calcular la sumatoria de los cuadrados de los valores después de restar la media e imprimir el resultado y las iteraciones para datos pares
    resultado_sumatoria_cuadrados_resta_media_par, iteraciones_sumatoria_cuadrados_resta_media_par = estadisticas_calculadora_par.calcular_sumatoria_cuadrados_resta_media()

    # Calcular la sumatoria de los cuadrados de los valores después de restar la media e imprimir el resultado y las iteraciones para datos impares
    resultado_sumatoria_cuadrados_resta_media_impar, iteraciones_sumatoria_cuadrados_resta_media_impar = estadisticas_calculadora_impar.calcular_sumatoria_cuadrados_resta_media()

    # Imprimir tabla combinada para datos pares
    imprimir_tabla_combinada(iteraciones_media_par, iteraciones_resta_media_par, iteraciones_sumatoria_cuadrados_par, iteraciones_sumatoria_cuadrados_resta_media_par, resultado_media_par)

    # Imprimir tabla combinada para datos impares
    imprimir_tabla_combinada(iteraciones_media_impar, iteraciones_resta_media_impar, iteraciones_sumatoria_cuadrados_impar, iteraciones_sumatoria_cuadrados_resta_media_impar, resultado_media_impar)

    # Crear una instancia de la clase DesviacionDeLaMedia para datos pares
    desviacion_calculadora_par = DesviacionDeLaMedia(estadisticas_calculadora_par)

    # Crear una instancia de la clase DesviacionDeLaMedia para datos impares
    desviacion_calculadora_impar = DesviacionDeLaMedia(estadisticas_calculadora_impar)

    # Calcular la desviación de la media para datos pares
    resultado_desviacion_media_par = desviacion_calculadora_par.calcular_desviacion_media()

    # Calcular la desviación de la media para datos impares
    resultado_desviacion_media_impar = desviacion_calculadora_impar.calcular_desviacion_media()

    # Imprimir el resultado para datos pares
    if resultado_desviacion_media_par is not None:
        print(f"La desviación de la media para datos pares es: {resultado_desviacion_media_par}")
    else:
        print("No hay suficientes datos para calcular la desviación de la media para datos pares.")

    # Imprimir el resultado para datos impares
    if resultado_desviacion_media_impar is not None:
        print(f"La desviación de la media para datos impares es: {resultado_desviacion_media_impar}")
    else:
        print("No hay suficientes datos para calcular la desviación de la media para datos impares.")

    # Crear una instancia de la clase DesviacionEstandar para datos pares
    desviacion_estandar_calculadora_par = DesviacionEstandar(desviacion_calculadora_par)

    # Crear una instancia de la clase DesviacionEstandar para datos impares
    desviacion_estandar_calculadora_impar = DesviacionEstandar(desviacion_calculadora_impar)

    # Calcular la desviación estándar para datos pares
    resultado_desviacion_estandar_par = desviacion_estandar_calculadora_par.calcular_desviacion_estandar()

    # Calcular la desviación estándar para datos impares
    resultado_desviacion_estandar_impar = desviacion_estandar_calculadora_impar.calcular_desviacion_estandar()

    # Imprimir los resultados para datos pares
    if resultado_desviacion_estandar_par is not None:
        print(f"La desviación estándar para datos pares es: {resultado_desviacion_estandar_par}")
    else:
        print("No hay suficientes datos para calcular la desviación estándar para datos pares.")

    # Imprimir los resultados para datos impares
    if resultado_desviacion_estandar_impar is not None:
        print(f"La desviación estándar para datos impares es: {resultado_desviacion_estandar_impar}")
    else:
        print("No hay suficientes datos para calcular la desviación estándar para datos impares.")

```
